```python
X = df.drop(columns= ["price"])
```

```python
y = df["price"]
```

```python
y_trans = np.log1p(y)
```

```python
columns_to_encode = ['property_type', 'sector','balcony',  
       'floortype', 'agePossession',  'furnishing_type', 'luxury_score']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
preprocessor = ColumnTransformer(transformers = [("num",StandardScaler(), ["bedRoom", "bathroom", "built_up_area", "store room", "servant room"]),
                                 ("cat", OrdinalEncoder(), columns_to_encode)], remainder = "passthrough")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
pipeline = Pipeline([("step1",preprocessor), ("step2", LinearRegression())
])
```

```python
pipeline
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score, KFold
kfold = KFold(n_splits= 10, shuffle= True, random_state= 42)
score = cross_val_score(pipeline, X, y_trans, cv = kfold, scoring= "r2")
```

```python
score.mean(), score.std()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_trans, random_state= 42 , test_size= 0.2)
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred= pipeline.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error

# Predicting on test set
y_pred = pipeline.predict(X_test)

# Calculate R-squared score
r2 = r2_score(np.expm1(y_test), np.expm1(y_pred))
print("R-squared score:", r2)

# Calculate Mean Absolute Error
mae = mean_absolute_error(np.expm1(y_test), np.expm1(y_pred))
print("Mean Absolute Error:", mae)
```

```python
def scorer(model_name, model):
    
    output = []
    
    output.append(model_name)
    
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])
    
    # K-fold cross-validation
    kfold = KFold(n_splits=10, shuffle=True, random_state=42)
    scores = cross_val_score(pipeline, X, y_trans, cv=kfold, scoring='r2')
    
    output.append(scores.mean())
    
    X_train, X_test, y_train, y_test = train_test_split(X,y_trans,test_size=0.2,random_state=42)
    
    pipeline.fit(X_train,y_train)
    
    y_pred = pipeline.predict(X_test)
    
    y_pred = np.expm1(y_pred)
    
    output.append(mean_absolute_error(np.expm1(y_test),y_pred))
    
    return output
    
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
model_dict = {
    'linear_reg':LinearRegression(),
    'svr':SVR(),
    'ridge':Ridge(),
    'LASSO':Lasso(),
    'decision tree': DecisionTreeRegressor(),
    'random forest':RandomForestRegressor(),
    'extra trees': ExtraTreesRegressor(),
    'gradient boosting': GradientBoostingRegressor(),
    'adaboost': AdaBoostRegressor(),
    'mlp': MLPRegressor(),
    'xgboost':XGBRegressor()
}
```

```python
model_output = []
for model_name,model in model_dict.items():
    model_output.append(scorer(model_name, model))
```
